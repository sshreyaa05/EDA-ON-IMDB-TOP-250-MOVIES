### as we can see that there are no missing values in the dataset , we can proceed further ....

### as we can see that under the column "budget" we have some values addressed as  "not available", therefore,  we can drop those particular rows with those " not available" values.......

```python
df_budget=df[df["budget"]=="Not Available"].index
df.drop(df_budget,inplace=True)
df.head(15)
```

## now lets convert the budget into numeric....

```python
df["budget"]=df["budget"].replace(" $"," ")
df["budget"]
```

```python
df[df["budget"]=="RF\xa018000000"]
```

```python
df["budget"]=df["budget"].str.replace("RF\xa0"," ")
df[df["budget"]=="EM\xa08000000"]
```

```python
df["budget"]=df["budget"].str.replace("EM\xa0"," ")
df["budget"]=df["budget"].str.replace("$"," ")
```

```python
df["budget"]=df["budget"].astype(float)
df["budget"]
```

```python
df.info()
```

## so lets convert the box office into numeric or float

```python
df[df["box_office"]=="Not Available"]
```

```python
df.transpose()
```

```python
df["box_office"]=df["box_office"].str.replace("(estimated)" , "")
df["box_office"]
```

```python
df["box_office"]= df["box_office"].replace("910000 ()" , "910000")
df["box_office"]
```

```python
df["box_office"]=df["box_office"].replace("750000 ()" , "750000")
df["box_office"]
```

```python
df["box_office"]= df["box_office"].replace("800000 ()" , "800000")
df["box_office"]
```

```python
df[df["box_office"]=="()"]
```

```python
df["box_office"]= df["box_office"].astype(float)
df["box_office"]
```

```python
df.info()
```

## now lets convert the run time into numeric

```python
indexrun_time=df[df["run_time"]=="Not Available"].index
df.drop(indexrun_time,inplace=True)
df.head(15)
```

```python
df[df["run_time"]=="Not Available"]
```

```python
df["run_time"]=df["run_time"].str.replace("h" , " ")
df["run_time"]
```

```python
df["run_time"]=df["run_time"].str.replace("m" , " ")
df["run_time"]
```

```python
df["run_time"]=df["run_time"].str.replace("  " , ".")
df["run_time"]
```

```python
df["run_time"]=df["run_time"].astype(float)
df["run_time"]
```

```python
df.info()
```

## 12. which director had the highest box office collection?

## 13. which genre has the highest box office collection and which year it had the highest box office collection?

```python
genre_high_boxoffice=df[["genre" , "box_office" , "year"]].groupby(["genre" , "box_office"]).sum().sort_values(by="box_office" , ascending=False)


genre_high_boxoffice
```

## result:
### so as we can see that genres "action" , "adventure" & "drama" has the highest box office collection in the year 2019 with 2.7994 billion dollars

## 14. which year has the highest box office collection with all the movies released in that year?
